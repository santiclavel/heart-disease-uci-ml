```python
# Heart Disease UCI - An谩lisis y Modelado en Google Colab

# Paso 1: Subir el archivo CSV desde tu computadora
from google.colab import files
import pandas as pd
import io

uploaded = files.upload()

# Detecta autom谩ticamente el nombre del archivo subido
filename = next(iter(uploaded))
print(f"Archivo subido: {filename}")

# Lee el archivo usando el nombre detectado
df = pd.read_csv(io.BytesIO(uploaded[filename]))

# Paso 2: Exploraci贸n del Dataset
print("Primeras filas del dataset:")
display(df.head())

print("\nInformaci贸n general del dataset:")
df.info()

print("\nDescripci贸n estad铆stica b谩sica:")
display(df.describe())

# Distribuci贸n de la variable objetivo
print("\nDistribuci贸n de la variable objetivo (target):")
print(df['target'].value_counts())

# Gr谩fico de caja (boxplot)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='chol')
plt.title("Boxplot de colesterol por clase de enfermedad card铆aca")
plt.show()

# Gr谩fico de viol铆n (violinplot)
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='target', y='chol')
plt.title("Violinplot de colesterol por clase de enfermedad card铆aca")
plt.show()

# Mapa de calor de correlaciones
plt.figure(figsize=(14,10))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Mapa de calor de correlaci贸n entre variables")
plt.show()

# Paso 3: Limpieza y preparaci贸n de datos
print("\nValores nulos por columna antes de limpiar:")
print(df.isnull().sum())

# Eliminamos filas con valores nulos
df = df.dropna()

print("\nDespu茅s de eliminar filas con NaN:")
print(df.isnull().sum())

# Codificaci贸n de variables categ贸ricas (si existen)
cat_cols = df.select_dtypes(include=['object']).columns
if len(cat_cols) > 0:
    print("\nColumnas categ贸ricas detectadas:", list(cat_cols))
    df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
else:
    print("\nNo se detectaron columnas categ贸ricas para codificar.")

# Paso 4: Selecci贸n de caracter铆sticas
# (En este ejemplo, usamos todas las variables excepto la target)
X = df.drop("target", axis=1)
y = df["target"]

# Paso 5: Escalado de variables num茅ricas
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Paso 6: Divisi贸n en conjunto de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Paso 7: Entrenamiento de modelos

# Modelo 1: Regresi贸n Log铆stica
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print(" Regresi贸n Log铆stica")
print(classification_report(y_test, y_pred_lr))

# Matriz de Confusi贸n
ConfusionMatrixDisplay.from_estimator(lr, X_test, y_test)
plt.title("Matriz de Confusi贸n - Regresi贸n Log铆stica")
plt.show()

# Modelo 2: Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print(" Random Forest")
print(classification_report(y_test, y_pred_rf))

# Matriz de Confusi贸n
ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)
plt.title("Matriz de Confusi贸n - Random Forest")
plt.show()

# Paso 8: Comparaci贸n de Rendimiento
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def metricas(nombre, y_true, y_pred):
    return {
        "Modelo": nombre,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1 Score": f1_score(y_true, y_pred)
    }

resultados = pd.DataFrame
```
